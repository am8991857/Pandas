## SECTION 4: Data Selection (loc & iloc)

### Quick Review

````python
# Create data with custom index
df_indexed = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}, index=['E001', 'E002', 'E003'])

print("DataFrame:")
print(df_indexed)

print("\nUsing .loc[]:")
print(df_indexed.loc['E002'])  # By label

print("\nUsing .iloc[]:")
print(df_indexed.iloc[1])  # By position
````

## SECTION 5: Modifying Data

## SECTION 7: Grouping & Aggregation

### Quick Review

````python
sales_data = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'South'],
    'Product': ['A', 'A', 'B', 'B'],
    'Amount': [100, 150, 200, 120]
})

print("Sales by Region:")
print(sales_data.groupby('Region')['Amount'].sum())

print("\nSales by Product:")
print(sales_data.groupby('Product')['Amount'].mean())
````

## SECTION 8: Combining DataFrames

### Quick Review

````python
# Concatenation (vertical stacking)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

combined = pd.concat([df1, df2], ignore_index=True)
print("Concatenated:")
print(combined)
````

````python
# Merging (join on key)
customers = pd.DataFrame({'CustID': [1, 2], 'Name': ['Alice', 'Bob']})
orders = pd.DataFrame({'CustID': [1, 2], 'Amount': [100, 200]})

merged = pd.merge(customers, orders, on='CustID')
print("\nMerged:")
print(merged)
````

## SECTION 10: Time Series

### Quick Review

````python
# Create time series
dates = pd.date_range('2024-01-01', periods=30, freq='D')
prices = pd.Series(np.random.randint(100, 150, 30), index=dates)

print("First 5 days:")
print(prices.head())

# Resample
weekly_avg = prices.resample('W').mean()
print("\nWeekly average:")
print(weekly_avg)
````

## SECTION 11: String Operations

````python
emails = pd.Series(
    ['alice@example.com', 'bob@test.org', 'charlie@example.com'])

print("Uppercase:")
print(emails.str.upper())

print("\nExtract domain:")
print(emails.str.split('@').str[1])

print("\nContains 'example':")
print(emails.str.contains('example'))
````
